```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data manually extracted from google search results.
data = {
    'Country': ['Portugal', 'Italy', 'Switzerland', 'UK', 'Canada', 'New Zealand', 'Benin', 'India', 'Fiji', 'Australia', 'United States', 'Barbados'],
    'Salt Intake (g/day)': [10.66, 9.82, 9.14, 8.06, 8.22, 8.58, 10.20, 9.08, 10.29, 9.04, 9.16, 6.75],
    'CKD Prevalence (%)': [20.9, 8.0, 10.0, 10.0, 7.19, 11.0, 16.10, 13.24, 13.8, 6.0, 14.0, 13.3]
}

df = pd.DataFrame(data)

# Create the scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(df['Salt Intake (g/day)'], df['CKD Prevalence (%)'])

# Add labels to each point
for i, country in enumerate(df['Country']):
    plt.annotate(country, (df['Salt Intake (g/day)'][i], df['CKD Prevalence (%)'][i]))

# Add title and labels
plt.title('Correlation between Salt Intake and CKD Prevalence')
plt.xlabel('Salt Intake (g/day)')
plt.ylabel('CKD Prevalence (%)')

# Add a trendline
z = np.polyfit(df['Salt Intake (g/day)'], df['CKD Prevalence (%)'], 1)
p = np.poly1d(z)
plt.plot(df['Salt Intake (g/day)'], p(df['Salt Intake (g/day)']), "r--")


# Save the plot
plt.savefig('salt_ckd_correlation.png')

# Display the plot
plt.show()

# Display the dataframe as a table to the user
print(df)
```
